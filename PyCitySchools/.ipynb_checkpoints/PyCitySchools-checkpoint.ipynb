```python
#import dependencies
import pandas as pd
```

```python
#define file paths
studentDataFile="students_complete.csv"
schoolDataFile="schools_complete.csv"
```

```python
#generate data frames based on file names
studentDf=pd.read_csv(studentDataFile)
schoolDf=pd.read_csv(schoolDataFile)
```

```python
#merge data sets based on school name
schoolDataComplete=pd.merge(studentDf, schoolDf, how="left", on=["school_name", "school_name"])
schoolDataComplete.head()
```

```python
schoolDataComplete.dtypes
```

```python
#find out how many unique values exist in the school name column to get number of schools
numSchools=schoolDataComplete["school_name"].nunique()
numSchools
```

```python
#find out how many students there are by using count. We do not use nunique here in case students have same name
numStudents=schoolDataComplete["student_name"].count()
numStudents
```

```python
#group the data by school name then use unique to get budget as it is listed multiple times
budgetDf=schoolDataComplete.groupby(["school_name"])
budgetSeries=budgetDf["budget"].unique()
#convert budget to a float so we can add it
budgetSeriesFloats=budgetSeries.astype(float)
totalBudget=budgetSeriesFloats.sum()
totalBudget
```

```python
#take mean of math score column
avgMathScore=schoolDataComplete["math_score"].mean()
avgMathScore
```

```python
#take mean of reading score column
avgReadingScore=schoolDataComplete["reading_score"].mean()
avgReadingScore
```

```python
#filter columns for rows in which only scores greater than or equal to 70 exist then count those scores, divide them by total num students to get percentage
passingMathScore=schoolDataComplete.loc[schoolDataComplete["math_score"]>=70, :]
passingMathStudent=passingMathScore["Student ID"].count()
passingMathPercent=(passingMathStudent/numStudents)*100
passingMathPercent
```

```python
passingReadingScore=schoolDataComplete.loc[schoolDataComplete["reading_score"]>=70,:]
passingReadingStudent=passingReadingScore["Student ID"].count()
passingReadingPercent=(passingReadingStudent/numStudents)*100
passingReadingPercent
```

```python
#perform filtering on both math and reading score >70 conditions
overallPassingScore=schoolDataComplete.loc[(schoolDataComplete["math_score"]>=70)&(schoolDataComplete["reading_score"]>=70), :]
overallPassingStudent=overallPassingScore["Student ID"].count()
overallPassingPercent=(overallPassingStudent/numStudents)*100
overallPassingPercent
```

```python
#format our data so it looks cleaner
district={"Total Schools":[numSchools], "Total Students":[numStudents], "Total Budget":[totalBudget], "Average Math Score":[avgMathScore], "Average Reading Score":[avgReadingScore], "% Passing Math":[passingMathPercent], "% Passing Reading":[passingReadingPercent], "% Overall Passing":[overallPassingPercent]}
districtDf=pd.DataFrame(district)
districtDf["Total Students"]=districtDf["Total Students"].map("{:,}".format)
districtDf["Total Budget"]=districtDf["Total Budget"].astype(float).map("${:,.2f}".format)
districtDf["Average Math Score"]=districtDf["Average Math Score"].astype(float).map("{:.2f}".format)
districtDf["Average Reading Score"]=districtDf["Average Reading Score"].astype(float).map("{:.2f}".format)
districtDf["% Passing Math"]=districtDf["% Passing Math"].astype(float).map("{:.2f}%".format)
districtDf["% Passing Reading"]=districtDf["% Passing Reading"].astype(float).map("{:.2f}%".format)
districtDf["% Overall Passing"]=districtDf["% Overall Passing"].astype(float).map("{:.2f}%".format)
districtDf
```

```python
#create a group based on school names
schoolNamesGroup=schoolDataComplete.groupby(["school_name"])
```

```python
#get school type from school type column based on our school name group. Rename series for use in later data frame
schoolTypes=schoolNamesGroup["type"].unique()
schoolTypes=schoolTypes.rename("School Type")
schoolTypes
```

```python
#get total number of students in each school based on the count of student IDs, rename series
totalStudents=schoolNamesGroup["Student ID"].count()
totalStudents=totalStudents.rename("Total Students")
totalStudents
```

```python
#get total budget for school by using the unique function as budget is repeated throughout data frame, so a sum function would be inaccurate here
totalSchoolBudget=schoolNamesGroup["budget"].unique()
#rename series and convert to float so we can use budget per student series
totalSchoolBudget=totalSchoolBudget.rename("Total School Budget").astype(float)
totalSchoolBudget
```

```python
#get mean math score based on school group, rename data frame
avgMathScore=schoolNamesGroup["math_score"].mean()
avgMathScore=avgMathScore.rename("Average Math Score")
avgMathScore
```

```python
#get average reading score based on school group, rename data frame
avgReadingScore=schoolNamesGroup["reading_score"].mean()
avgReadingScore=avgReadingScore.rename("Average Reading Score")
avgReadingScore
```

```python
#create a lambda function where we count every score greater than or equal to 70. Use apply function to apply to every row of the data frame.
passingMathNum=schoolNamesGroup["math_score"].apply(lambda x: (x>=70).sum())
passingMathNum
```

```python
#count number of math scores by school to return total number of people taking math. 
#We then divide the number of people passing math by the number taking math to return the percentage of students patssing math
totalTakingMath=schoolNamesGroup["math_score"].count()
pctPassingMath=((passingMathNum/totalTakingMath)*100).rename("% Passing Math")
pctPassingMath
```

```python
#divide total school budget by total students to return the total per-student budget
perStudentBudget=(totalSchoolBudget/totalStudents).astype(float)
perStudentBudget=perStudentBudget.rename("Per Student Budget")
perStudentBudget
```

```python
#use lambda function to count reading scores greater than or equal to 70
#use apply function to apply to every row in data frame
passingReadingNum=schoolNamesGroup["reading_score"].apply(lambda x: (x>=70).sum())
passingReadingNum
```

```python
#count number of names in reading score series to return total students taking reading
#divide number of students passing reading by the number taking to return percentage of those passing
numTakingReading=schoolNamesGroup["reading_score"].count()
pctPassingReading=((passingReadingNum/numTakingReading)*100).rename("% Passing Reading")
pctPassingReading
```

```python
#take overall passing score determined in first half of exercise and group by school name
overallPassingScoreGroup=overallPassingScore.groupby(["school_name"])
#count the number of student IDs in the passing score group to determine number of students passing
overallPassingNum=overallPassingScoreGroup['Student ID'].count()
#divide the number of passing students by number of total students to return percentage of overall passing
passingBothPercent=((overallPassingNum/totalStudents)*100).rename("% Overall Passing")
passingBothPercent
```

```python
#concatenate the data frames. Here we do not use the merge function, as merge will only take two series. 
schoolSummaryDf=pd.concat([schoolTypes, totalStudents, totalSchoolBudget, perStudentBudget, avgMathScore, avgReadingScore, pctPassingMath, pctPassingReading, passingBothPercent], axis='columns', sort=False)
#apply formatting
# schoolSummaryDf["Total School Budget"]=schoolSummaryDf["Total School Budget"].map("${:,.2f}".format)
# schoolSummaryDf["Per Student Budget"]=schoolSummaryDf["Per Student Budget"].map("${:,.2f}".format)
schoolSummaryDf
```

```python
#sort dataframe based on top 5 academically performing schools
schoolSummaryDf.sort_values(by=["% Overall Passing"], ascending=False).head()
```

```python
#sort dataframe best on lowest 5 academically performing schools
schoolSummaryDf.sort_values(by=["% Overall Passing"], ascending=True).head()
```

```python
#specify which columns of data we would like to return for our series
columns=["grade", "school_name", "math_score"]
#filter data frame to return only those values that are associated with 9th grade
ninthGradeSeries=schoolDataComplete.loc[schoolDataComplete["grade"]=="9th", columns]
ninthGradeSeries.head()
```

```python
#filter data frame to return only those values that are associated with 10th grade
tenthGradeSeries=schoolDataComplete.loc[schoolDataComplete["grade"]=="10th", columns]
tenthGradeSeries.head()
```

```python
#filter data frame to return only those values that are associated with 11th grade
eleventhGradeSeries=schoolDataComplete.loc[schoolDataComplete["grade"]=="11th", columns]
eleventhGradeSeries.head()
```

```python
#filter data frame to return only those values that are associated with 12th grade
twelfthGradeSeries=schoolDataComplete.loc[schoolDataComplete["grade"]=="12th", columns]
twelfthGradeSeries.head()
```

```python
#get mean of 9th grade series math scores
ninthGradeSeriesGroup=ninthGradeSeries.groupby(["school_name"])
ninthGradeGroup=(ninthGradeSeriesGroup["math_score"].mean()).rename("9th")
ninthGradeGroup
```

```python
#get mean of 10th grade series math scores
tenthGradeSeriesGroup=tenthGradeSeries.groupby(["school_name"])
tenthGradeGroup=(tenthGradeSeriesGroup["math_score"].mean()).rename("10th")
tenthGradeGroup
```

```python
#get mean of 11th grade series math scores
eleventhGradeSeriesGroup=eleventhGradeSeries.groupby(["school_name"])
eleventhGradeGroup=(eleventhGradeSeriesGroup["math_score"].mean()).rename("11th")
eleventhGradeGroup
```

```python
#get mean of 12th grade series math scores
twelfthGradeSeriesGroup=twelfthGradeSeries.groupby(["school_name"])
twelfthGradeGroup=(twelfthGradeSeriesGroup["math_score"].mean()).rename("12th")
twelfthGradeGroup
```

```python
#combine series to form summary data frame by using concat function
mathScoresByGradeDf=pd.concat([ninthGradeGroup, tenthGradeGroup, eleventhGradeGroup, twelfthGradeGroup], axis='columns', sort=False)
mathScoresByGradeDf
```

```python
#perform same operations as above, using the reading score column 
columns2=columns=["grade", "school_name", "reading_score"]
ninthGradeReadingSeriesGroup=(schoolDataComplete.loc[schoolDataComplete["grade"]=="9th", columns2]).groupby(["school_name"])
tenthGradeReadingSeriesGroup=(schoolDataComplete.loc[schoolDataComplete["grade"]=="10th", columns2]).groupby(["school_name"])
eleventhGradeReadingSeriesGroup=(schoolDataComplete.loc[schoolDataComplete["grade"]=="11th", columns2]).groupby(["school_name"])
twelfthGradeReadingSeriesGroup=(schoolDataComplete.loc[schoolDataComplete["grade"]=="12th", columns2]).groupby(["school_name"])
ninthGradeReadingGroup=(ninthGradeReadingSeriesGroup["reading_score"].mean()).rename("9th")
tenthGradeReadingGroup=(tenthGradeReadingSeriesGroup["reading_score"].mean()).rename("10th")
eleventhGradeReadingGroup=(eleventhGradeReadingSeriesGroup["reading_score"].mean()).rename("11th")
twelfthGradeReadingGroup=(twelfthGradeReadingSeriesGroup["reading_score"].mean()).rename("12th")
```

```python
#combine series into reading score summary data frame
readingScoresByGradeDf=pd.concat([ninthGradeReadingGroup, tenthGradeReadingGroup, eleventhGradeReadingGroup, twelfthGradeReadingGroup], axis='columns', sort=False)
readingScoresByGradeDf
```
